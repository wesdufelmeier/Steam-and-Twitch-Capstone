```python
import json
import time
from pathlib import Path
import steamspypi
import pandas as pd
```

```python
def get_cooldown():
    cooldown = 70  # 1 minute plus a cushion

    return cooldown


def get_some_sleep():
    cooldown = get_cooldown()
    print("Sleeping for {} seconds on {}".format(cooldown, time.asctime()))

    time.sleep(cooldown)

    return


def download_a_single_page(page_no=0):
    print("Downloading page={} on {}".format(page_no, time.asctime()))

    data_request = dict()
    data_request["request"] = "all"
    data_request["page"] = str(page_no)

    data = steamspypi.download(data_request)

    return data


def get_file_name(page_no):
    # Get current day as yyyymmdd format
    date_format = "%Y%m%d"
    current_date = time.strftime(date_format)
    
    file_name = "{}_steamspy_page_{}.json".format(current_date, page_no)
    
    return file_name


def download_all_pages(num_pages):
    # Download

    for page_no in range(num_pages):
        file_name = get_file_name(page_no)

        if not Path(file_name).is_file():
            page_data = download_a_single_page(page_no=page_no)

            with open(file_name, "w", encoding="utf8") as f:
                json.dump(page_data, f)

            if page_no != (num_pages - 1):
                get_some_sleep()

    # Aggregate

    data = dict()

    for page_no in range(num_pages):
        file_name = get_file_name(page_no)

        with open(file_name, "r", encoding="utf8") as f:
            page_data = json.load(f)

            data.update(page_data)

    return data


if __name__ == "__main__":
    # TODO: one would have to figure out the number of pages, it should be close to 40 as of August 2020.
    data = download_all_pages(num_pages=80)
```

```python
from tqdm import tqdm  # Progress bar
import os  # To check if the file already exists and handle appending

# List of app IDs
app_ids 

# Checkpoint filename
checkpoint_file = 'steam_app_data.csv'

# Check if the checkpoint file already exists
if os.path.exists(checkpoint_file):
    # If it exists, load the existing data and resume from the last processed app ID
    existing_df = pd.read_csv(checkpoint_file)
    existing_app_ids = existing_df['appid'].tolist()  # Assuming 'appid' is a column in the CSV
    start_index = len(existing_app_ids)
    results = existing_df.to_dict(orient='records')  # Convert existing data to a list of dictionaries
    print(f"Resuming from app ID index {start_index} ({app_ids[start_index] if start_index < len(app_ids) else 'None'})")
else:
    # If the file doesn't exist, start from scratch
    start_index = 0
    results = []

# Loop over each app ID, starting from where the last checkpoint ended
for i, app_id in enumerate(tqdm(app_ids[start_index:], desc="Downloading Apps", unit="app", ncols=100)):
    data_request = dict()
    data_request['request'] = 'appdetails'
    data_request['appid'] = str(app_id)  # Ensure app_id is a string
    
    # Download the data
    try:
        data = steamspypi.download(data_request)
        # Append the data to the results list
        results.append(data)
        
        # Every 1000 iterations, save a checkpoint to the CSV file
        if (i + start_index + 1) % 1000 == 0:
            temp_df = pd.DataFrame(results)
            temp_df.to_csv(checkpoint_file, index=False)
            print(f"Checkpoint saved at iteration {i + start_index + 1}")
    except Exception as e:
        print(f"Error downloading data for app {app_id}: {e}")
        continue

# Final saving of all results to the CSV
df = pd.DataFrame(results)
df.to_csv(checkpoint_file, index=False)

print(f"Data saved to {checkpoint_file}")
```
